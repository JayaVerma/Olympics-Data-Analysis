## 2. Which sport is having most number of Gold Medals so far? (Top 5)¶

```python
medal = data['Medal'].unique()
medal
```

```python
sports = data['Sport'].unique()
sports
```

```python
len(sports)
```

```python
GoldMedal = data[data['Medal']=='Gold'].groupby(['Sport'])['Medal'].count()
GoldMedal
```

```python
goldMedal = GoldMedal.reset_index(name='Gold Medal').sort_values(by='Gold Medal', ascending=False)
goldMedal
```

```python
mostGoldMedal = goldMedal.head()
mostGoldMedal.set_index('Sport',inplace=True)
mostGoldMedal
```

```python
font1 = {'family':'serif', 'color':'blue', 'size':25}
font2 = {'family':'serif', 'color':'red', 'size':15}

mostGoldMedal.plot.bar(figsize=(12,6))
plt.title('Top 5 Sport having most number of Gold Medals', fontdict=font1)
plt.xlabel('Sport', fontdict=font2)
plt.ylabel("Number of Gold Medels", fontdict=font2)
```

## 3. Which sport is having most number of medals so far? (Top 5)¶

```python
Medal = data['Sport'].value_counts().reset_index()
Medal.columns=['Sport','No of Medals']
Medal
```

```python
mostMedal = Medal.head()
mostMedal.set_index('Sport',inplace=True)
mostMedal
```

```python
font1 = {'family':'serif', 'color':'blue', 'size':25}
font2 = {'family':'serif', 'color':'red', 'size':15}

mostMedal.plot.bar(figsize=(12,6))
plt.title('Top 5 Sport having most number of Medals', fontdict=font1)
plt.xlabel('Sport', fontdict=font2)
plt.ylabel("Number of Medels", fontdict=font2)
```

# 4. Which player has won most number of medals? (Top 5)

```python
player = data['Athlete'].value_counts().reset_index()
player.columns=['Athlete','No of Medals']
player
```

```python
playerMedal = player.head()
playerMedal.set_index('Athlete',inplace=True)
playerMedal
```

```python
font1 = {'family':'serif', 'color':'blue', 'size':25}
font2 = {'family':'serif', 'color':'red', 'size':15}

playerMedal.plot.bar(figsize=(12,6))
plt.title('Top 5 Player won most number of Medals', fontdict=font1)
plt.xlabel('Athlete', fontdict=font2)
plt.ylabel("Number of Medels Won", fontdict=font2)
```

## 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
playerGoldMedal = data[data['Medal']=='Gold'].groupby(['Athlete'])['Medal'].count().reset_index(name='Gold Medals').sort_values(by='Gold Medals',ascending=False).head()
playerGoldMedal.set_index('Athlete',inplace=True)
playerGoldMedal
```

```python
font1 = {'family':'serif', 'color':'blue', 'size':25}
font2 = {'family':'serif', 'color':'red', 'size':15}

playerGoldMedal.plot.bar(figsize=(12,6))
plt.title('Top 5 Player won most number of Gold Medals', fontdict=font1)
plt.xlabel('Athlete', fontdict=font2)
plt.ylabel("Number of Gold Medels Won", fontdict=font2)
```

## 7. Which event is most popular in terms on number of players? (Top 5)

```python
event = data['Event'].value_counts().reset_index()
event.columns=['Event','Players']
event
```

```python
topEvent = event.head()
topEvent.set_index('Event',inplace=True)
topEvent
```

```python
font1 = {'family':'serif', 'color':'blue', 'size':25}
font2 = {'family':'serif', 'color':'red', 'size':15}

topEvent.plot.bar(figsize=(12,6))
plt.title('Top 5 most popular Events in terms of No. of Players', fontdict=font1)
plt.xlabel('Event', fontdict=font2)
plt.ylabel("Number of Players", fontdict=font2)
```

## 8. Which sport is having most female Gold Medalists? (Top 5)

```python
women = data[data['Gender']=='Women']
GoldMedalist = women[women['Medal']=='Gold']
GoldMedalist
```

```python
womenGold = GoldMedalist['Sport'].value_counts().reset_index()
womenGold.columns=['Sport','No. of Women Players']
womenGold
```

```python
womenGoldMedalist = womenGold.head()
womenGoldMedalist.set_index('Sport',inplace=True)
womenGoldMedalist
```

```python
font1 = {'family':'serif', 'color':'blue', 'size':25}
font2 = {'family':'serif', 'color':'red', 'size':15}

womenGoldMedalist.plot.bar(figsize=(12,6))
plt.title('Top 5 Sports having most femal gold Medalist', fontdict=font1)
plt.xlabel('Sport', fontdict=font2)
plt.ylabel("No. of women Gold Medalist", fontdict=font2)
```
